```python
def find_string_in_df(data_frame, partial_string1, partial_string2 = None):
    # Check if each element in the DataFrame contains the first partial string
    matches1 = data_frame.apply(lambda col: col.astype(str).str.contains(partial_string1, case=True))

    # Get the row and column indices where the first partial string matches
    rows1, cols1 = matches1.values.nonzero()

    # If the first string is found, return the result
    if len(rows1) > 0:
        return rows1[0], cols1[0], data_frame.columns[cols1[0]]

    # If a second partial string is provided, check for it
    if partial_string2 is not None:
        matches2 = data_frame.apply(lambda col: col.astype(str).str.contains(partial_string2, case=True))
        rows2, cols2 = matches2.values.nonzero()

        # If the second string is found, return the result
        if len(rows2) > 0:
            return rows2[0], cols2[0], data_frame.columns[cols2[0]]

    # Return None if neither string is found
    return None
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os


def search_text_in_df(dataframe, text):
    for col in dataframe.columns:

        rows = dataframe[col].apply(lambda x: str(x).lower() == text.lower()).any()
        if rows:
            found_row = dataframe[dataframe[col].apply(lambda x: str(x).lower() == text.lower())].index.tolist()
            if found_row:
                return found_row[0], col
    return None, None


url = 'https://www.dca.ga.gov/node/4696'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

download_links = soup.find_all('a', href=True)
excel_links1 = [link['href'] for link in download_links if link['href'].endswith('.xls')]


search_text = "waiver of qualified contract right or tenant ownership plan"


dfs = []


for link in excel_links1:

    if not link.startswith('http'):
        link = 'BASE_URL' + link

    file_name = link.split('/')[-1]
    file_path = '/content/' + file_name


    if not os.path.exists(file_path):
        print(f"Downloading file: {file_name}")
        response = requests.get(link)
        with open(file_path, 'wb') as file:
            file.write(response.content)
    else:
        print(f"File already downloaded: {file_name}")


    xls = pd.ExcelFile(file_path)
    sheet_names = xls.sheet_names


    for sheet in sheet_names:
        df_sheet = pd.read_excel(xls, sheet_name=sheet)
        found_row, found_col = search_text_in_df(df_sheet, search_text)
        if found_row is not None:
            break

    if found_row is None:
        print(f"Text not found in {file_name}")
        continue


    property_name = file_name.split('_')[1]


    table = pd.DataFrame(columns=['Column 1', 'Column 2'], index=range(1, 5))
    table.at[1, 'Column 1'] = property_name
    table.at[1, 'Column 2'] = ""
    table.at[2, 'Column 1'] = search_text
    table.at[2, 'Column 2'] = df_sheet.iloc[found_row, df_sheet.columns.get_loc(found_col) + 13] if df_sheet.shape[1] > df_sheet.columns.get_loc(found_col) + 13 else "Out of range"
    table.at[3, 'Column 1'] = df_sheet.iloc[found_row + 2, df_sheet.columns.get_loc(found_col) + 1] if df_sheet.shape[0] > found_row + 2 else "Out of range"
    table.at[3, 'Column 2'] = df_sheet.iloc[found_row + 2, df_sheet.columns.get_loc(found_col) + 13] if df_sheet.shape[1] > df_sheet.columns.get_loc(found_col) + 13 else "Out of range"
    table.at[4, 'Column 1'] = df_sheet.iloc[found_row + 3, df_sheet.columns.get_loc(found_col) + 1] if df_sheet.shape[0] > found_row + 3 else "Out of range"
    table.at[4, 'Column 2'] = df_sheet.iloc[found_row + 3, df_sheet.columns.get_loc(found_col) + 13] if df_sheet.shape[1] > df_sheet.columns.get_loc(found_col) + 13 else "Out of range"


    dfs.append(table)


yes_forgo = no_forgo = nan_forgo = 0
yes_ownership = no_ownership = nan_ownership = 0


for df in dfs:

    response_forgo = df.at[3, 'Column 2']
    if response_forgo == 'Yes':
        yes_forgo += 1
    elif response_forgo == 'No':
        no_forgo += 1
    elif pd.isna(response_forgo):
        nan_forgo += 1


    response_ownership = df.at[4, 'Column 2']
    if response_ownership == 'Yes':
        yes_ownership += 1
    elif response_ownership == 'No':
        no_ownership += 1
    elif pd.isna(response_ownership):
        nan_ownership += 1


summary_data = {
    "# of Yes": [yes_forgo, yes_ownership],
    "# of No": [no_forgo, no_ownership],
    "# of nans": [nan_forgo, nan_ownership]
}

summary_df = pd.DataFrame(summary_data, index=[
    "Willing to forgo cancellation option after the close of the compliance period of 5 years",
    "Tenant Ownership Plans"
])

summary_df.name = "Total Values"
print(summary_df)
```
